```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 3)
y = np.linspace(-3, 3, 4)

(X, Y) = np.meshgrid(x, y)
R = np.sqrt(X**2+Y**2)

plt.figure()
CS = plt.contour(X, Y, R)
plt.clabel(CS, inline=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import ma

'''
Calculate Electric Potential for a Point Charge
'''
def Electric_Potential_Point(Distance_X=1,Distance_Y=1,Point_Charge=1.602e-19,K=8.9875517873681764e9, White_Noise=1e-12):
    Distance_R=np.sqrt(Distance_X*Distance_X+Distance_Y*Distance_Y)
# Replace small distances < White Noise
    np.clip(Distance_R,White_Noise,None,Distance_R)
    Electric_Potential=K*Point_Charge/Distance_R
    return Electric_Potential
    
Q_Positive=1.602e-19
Q_Negative=-Q_Positive
(Point_Positive_X, Point_Positive_Y)=(0,0)

X, Y = np.meshgrid(np.arange(-1e-11,1e-11,1e-13), np.arange(-1e-11,1e-11,1e-13))

Distance_Positive_X=X-Point_Positive_X
Distance_Positive_Y=Y-Point_Positive_Y

Electric_Potential = Electric_Potential_Point(Distance_Positive_X, Distance_Positive_Y,Q_Negative)

plt.figure()
plt.title('Electric Potential for a Point Charge')
CS=plt.contour(X, Y, Electric_Potential)
plt.clabel(CS, inline=True, fontsize=8)
CB=plt.colorbar(CS, shrink=0.8, extend='both')
plt.axis('equal')
plt.show()
```
